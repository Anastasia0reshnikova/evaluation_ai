```python
# Импортируем библиотеки - это как подключение инструментов для работы
import pandas as pd  # Для работы с таблицами данных
from sklearn.model_selection import train_test_split  # Функция для разделения данных
```

```python
# pd.read_csv() - функция читает CSV-файл и создает DataFrame. Возвращает табличную структуру данных.
df = pd.read_csv('airline_passenger_satisfaction.csv')
# pd.set_option() - функция устанавливает параметры отображения pandas. 'display.max_columns', None означает показывать все столбцы без ограничений.
pd.set_option('display.max_columns', None)
# head() - метод возвращает первые 5 строк DataFrame (по умолчанию), полезен для быстрого просмотра структуры данных.
df.head()
```

```python
# drop() - метод удаляет указанные столбцы или строки из DataFrame

# columns - список столбцов для удаления
# inplace=True - изменяет исходный DataFrame без создания копии

df.drop(columns = ['id', 'Unnamed: 0'], inplace=True)
df.dropna(inplace=True)
df.head()
```

```python
# df.drop(columns='satisfaction') - удаляет столбец 'satisfaction', создавая матрицу признаков X

X = df.drop("satisfaction", axis=1)
X.head()
```

```python
y = df['satisfaction'] # Извлекает целевую переменную (то, что мы хотим предсказать)
y.head()
```

```python
# Преобразуем текстовые признаки в числовые
X = pd.get_dummies(X, drop_first=True)
```

```python
# ПРАВИЛЬНОЕ разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X, y,                    # Наши данные
    test_size=0.2,          # 20% данных для тестирования
    random_state=42,        # Фиксируем случайность для воспроизводимости
    stratify=y              # Сохраняем пропорции классов в обеих выборках
)
```

### Random Forest (Домашнее задание)

```python
from sklearn.ensemble import RandomForestClassifier # RandomForestClassifier - алгоритм "случайный лес", который объединяет множество деревьев решений
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score, confusion_matrix  # Импортируем метрики точности

model_random_forest = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
print("Начинаем обучение модели RandomForestClassifier")

# Обучаем модель
model_random_forest.fit(X_train, y_train)

print("Модель RandomForestClassifier обучена")
```

```python
# Делаем предсказания
predictions_rf = model_random_forest.predict(X_test)
```

```python
# Вычисление всех метрик качества модели
accuracy = accuracy_score(y_test, predictions_rf)  # accuracy_score() - сравнивает истинные значения y_test с предсказаниями и возвращает долю совпадений
# pos_label='satisfied' → считаем метрику только для одного класса. Только для бинарной классификации (satisfied vs neutral or dissatisfied).
# В таком режиме метрика отвечает на вопрос: «Насколько хорошо модель умеет находить именно класс 'satisfied'?»
precision = precision_score(y_test, predictions_rf, pos_label='satisfied')  # precision_score() - вычисляет точность (из предсказанных положительных, сколько действительно положительных)
recall = recall_score(y_test, predictions_rf, pos_label='satisfied')  # recall_score() - вычисляет полноту (из всех положительных случаев, сколько мы нашли)
f1 = f1_score(y_test, predictions_rf, pos_label='satisfied')  # f1_score() - гармоническое среднее между precision и recall

# average='weighted' → берём оба класса и усредняем их результат, учитывая их количество.
# average='weighted' отвечает на вопрос: «В среднем по всем классам, насколько хорошо работает модель»
f_beta_05 = fbeta_score(y_test, predictions_rf, beta=0.5, average='weighted')  # fbeta_score() с beta=0.5 - больше веса precision
f_beta_2 = fbeta_score(y_test, predictions_rf, beta=2.0, average='weighted')  # fbeta_score() с beta=2.0 - больше веса recall

print("📊 РЕЗУЛЬТАТЫ ОЦЕНКИ МОДЕЛИ RandomForestClassifier:")
print("="*50)
print(f"Accuracy (Точность):           {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f"Precision (Точность полож.):   {precision:.4f} ({precision*100:.2f}%)")
print(f"Recall (Полнота):              {recall:.4f} ({recall*100:.2f}%)")
print(f"F1-Score:                      {f1:.4f} ({f1*100:.2f}%)")
print(f"F-Beta (β=0.5, упор на Precis): {f_beta_05:.4f} ({f_beta_05*100:.2f}%)")
print(f"F-Beta (β=2.0, упор на Recall): {f_beta_2:.4f} ({f_beta_2*100:.2f}%)")

print(f"\n📝 Интерпретация результатов:")
print(f"• Из {len(y_test)} пассажиров модель правильно классифицировала {int(accuracy * len(y_test))}")
print(f"• Общая точность составляет {accuracy*100:.2f}%")
```

### KNeighborsClassifier (Домашнее задание)

```python
from sklearn.neighbors import KNeighborsClassifier # KNeighborsClassifier - алгоритм k ближайших соседей n_neighbors=5 - количество соседей для классификации
from sklearn.preprocessing import StandardScaler

model_kn_neighbors_classifier = KNeighborsClassifier(
    n_neighbors=5
)
print("Начинаем обучение модели KNeighborsClassifier")

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Обучаем модель
model_kn_neighbors_classifier.fit(X_train_scaled, y_train)
print("Модель KNeighborsClassifier обучена")
```

```python
# Делаем предсказания
predictions_knn = model_kn_neighbors_classifier.predict(X_test_scaled)
```

```python
# Вычисление всех метрик качества модели
accuracy = accuracy_score(y_test, predictions_knn)  # accuracy_score() - сравнивает истинные значения y_test с предсказаниями и возвращает долю совпадений
# pos_label='satisfied' → считаем метрику только для одного класса. Только для бинарной классификации (satisfied vs neutral or dissatisfied).
# В таком режиме метрика отвечает на вопрос: «Насколько хорошо модель умеет находить именно класс 'satisfied'?»
precision = precision_score(y_test, predictions_knn, pos_label='satisfied')  # precision_score() - вычисляет точность (из предсказанных положительных, сколько действительно положительных)
recall = recall_score(y_test, predictions_knn, pos_label='satisfied')  # recall_score() - вычисляет полноту (из всех положительных случаев, сколько мы нашли)
f1 = f1_score(y_test, predictions_knn, pos_label='satisfied')  # f1_score() - гармоническое среднее между precision и recall

# average='weighted' → берём оба класса и усредняем их результат, учитывая их количество.
# average='weighted' отвечает на вопрос: «В среднем по всем классам, насколько хорошо работает модель»
f_beta_05 = fbeta_score(y_test, predictions_knn, beta=0.5, average='weighted')  # fbeta_score() с beta=0.5 - больше веса precision
f_beta_2 = fbeta_score(y_test, predictions_knn, beta=2.0, average='weighted')  # fbeta_score() с beta=2.0 - больше веса recall

print("📊 РЕЗУЛЬТАТЫ ОЦЕНКИ МОДЕЛИ KNeighborsClassifier:")
print("="*50)
print(f"Accuracy (Точность):           {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f"Precision (Точность полож.):   {precision:.4f} ({precision*100:.2f}%)")
print(f"Recall (Полнота):              {recall:.4f} ({recall*100:.2f}%)")
print(f"F1-Score:                      {f1:.4f} ({f1*100:.2f}%)")
print(f"F-Beta (β=0.5, упор на Precis): {f_beta_05:.4f} ({f_beta_05*100:.2f}%)")
print(f"F-Beta (β=2.0, упор на Recall): {f_beta_2:.4f} ({f_beta_2*100:.2f}%)")

print(f"\n📝 Интерпретация результатов:")
print(f"• Из {len(y_test)} пассажиров модель правильно классифицировала {int(accuracy * len(y_test))}")
print(f"• Общая точность составляет {accuracy*100:.2f}%")
```

### Logistic Regression (Домашнее задание)

```python
from sklearn.linear_model import LogisticRegression # LogisticRegression - алгоритм логистической регрессии max_iter=10000 - максимальное количество итераций для сходимости

model_logistic_regression = LogisticRegression(
    max_iter=10000
)
print("Начинаем обучение модели LogisticRegression")

# Обучаем модель
model_logistic_regression.fit(X_train_scaled, y_train)

print("Модель LogisticRegression обучена")
```

```python
# Делаем предсказания
predictions_lr = model_logistic_regression.predict(X_test_scaled)
```

## Сравнение результатов моделей

```python
results = pd.DataFrame({
    "Model": [
        "RandomForestClassifier",
        "KNeighborsClassifier",
        "LogisticRegression"
    ],
    "Accuracy": [
        accuracy_score(y_test, predictions_rf),
        accuracy_score(y_test, predictions_knn),
        accuracy_score(y_test, predictions_lr)
    ],
    "Precision": [
        precision_score(y_test, predictions_rf, pos_label="satisfied"),
        precision_score(y_test, predictions_knn, pos_label="satisfied"),
        precision_score(y_test, predictions_lr, pos_label="satisfied")
    ],
    "Recall": [
        recall_score(y_test, predictions_rf, pos_label="satisfied"),
        recall_score(y_test, predictions_knn, pos_label="satisfied"),
        recall_score(y_test, predictions_lr, pos_label="satisfied")
    ],
    "F1-score": [
        f1_score(y_test, predictions_rf, pos_label="satisfied"),
        f1_score(y_test, predictions_knn, pos_label="satisfied"),
        f1_score(y_test, predictions_lr, pos_label="satisfied")
    ]
})

results.sort_values(by="Accuracy", ascending=False)
```
