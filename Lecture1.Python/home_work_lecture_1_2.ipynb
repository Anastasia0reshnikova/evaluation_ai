```python
import csv

local_filename = "home_experiments.csv"
report_filename = "report.txt"

experiment_data = [
    ["experiment_id", "model_name", "accuracy", "training_time", "dataset_size"],
    ["exp_001", "BERT-small", 0.847, 45.2, 1000],
    ["exp_002", "GPT-2", 0.923, 120.8, 1000],
    ["exp_003", "BERT-large", 0.891, 180.5, 1000],
    ["exp_004", "Random-Forest", 0.756, 12.1, 1000],
    ["exp_005", "BERT-small", 0.852, 44.8, 2000],
    ["exp_006", "GPT-2", 0.945, 125.3, 2000],
    ["exp_007", "BERT-large", 0.912, 182.1, 2000],
    ["exp_008", "Random-Forest", 0.778, 23.4, 2000],
]

# 1. Создать CSV-файл с данными
with open(local_filename, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(experiment_data)

print(f"Данные сохранены в {local_filename}")


# 2. Написать функцию read_experiments(filename)
def read_experiments(filename: str):
    experiments = []

    with open(filename, "r", encoding="utf-8") as csvfile:
        reader = csv.DictReader(csvfile)

        for row in reader:
            experiment = {
                "experiment_id": row["experiment_id"],
                "model_name": row["model_name"],
                "accuracy": float(row["accuracy"]),
                "training_time": float(row["training_time"]),
                "dataset_size": int(row["dataset_size"]),
            }

            experiments.append(experiment)

    return experiments



experiments = read_experiments(local_filename)


print(f"\nЗагружено {len(experiments)} экспериментов:")
for exp in experiments:
    print(
        f"{exp['experiment_id']}: {exp['model_name']} - "
        f"точность {exp['accuracy']:.3f}"
    )


# 3. Создать список экспериментов с точностью выше 0.9
high_accuracy_experiments = []

for exp in experiments:
    if exp["accuracy"] > 0.9:
        high_accuracy_experiments.append(exp)

print("\nЭксперименты с точностью выше 0.9:")
for exp in high_accuracy_experiments:
    print(f"{exp['experiment_id']}: {exp['model_name']} - {exp['accuracy']}")


# 4. Найти все эксперименты с BERT-моделями
bert_experiments = []

for exp in experiments:
    if "BERT" in exp["model_name"]:
        bert_experiments.append(exp)

print("\nЭксперименты с BERT-моделями:")
for exp in bert_experiments:
    print(f"{exp['experiment_id']}: {exp['model_name']}")


# 5. Создать функцию, которая сохраняет сводный отчет в report.txt
def save_report(filename: str, experiments: list, high_accuracy_experiments: list, bert_experiments: list) -> None:
    with open(filename, "w", encoding="utf-8") as file:
        file.write("Сводный отчет по экспериментам\n")
        file.write("=" * 35 + "\n\n")

        file.write(f"Общее количество экспериментов: {len(experiments)}\n")
        file.write(
            f"Количество экспериментов с точностью выше 0.9: "
            f"{len(high_accuracy_experiments)}\n"
        )
        file.write(
            f"Количество экспериментов с BERT-моделями: "
            f"{len(bert_experiments)}\n\n"
        )

        file.write("Эксперименты с точностью выше 0.9:\n")
        for exp in high_accuracy_experiments:
            file.write(
                f"{exp['experiment_id']}: {exp['model_name']}, "
                f"accuracy={exp['accuracy']}, "
                f"training_time={exp['training_time']}, "
                f"dataset_size={exp['dataset_size']}\n"
            )

        file.write("\nЭксперименты с BERT-моделями:\n")
        for exp in bert_experiments:
            file.write(
                f"{exp['experiment_id']}: {exp['model_name']}, "
                f"accuracy={exp['accuracy']}, "
                f"training_time={exp['training_time']}, "
                f"dataset_size={exp['dataset_size']}\n"
            )


save_report("report.txt", experiments, high_accuracy_experiments, bert_experiments)

print("\nОтчет сохранен в report.txt")

```
